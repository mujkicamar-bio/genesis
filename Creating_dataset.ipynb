```python
def refseq_decider(df):
    refseq_to_download = {}

    for species, ref_id in zip(df["host_species"], df["plasmid_id"]):
        # Create list of refeseqs for download
            refseq_to_download[species] = []

        refseq_to_download[species].append(ref_id)

    return refseq_to_download
```

```python
blast = pd.read_csv("/Users/amamu656/Desktop/GENESIS/blast/results.tsv", sep="\t", names=["qseqid", "sseqid", "pident", "length",
                           "qstart", "qend", "sstart", "send", "evalue", "bitscore"])
blast
```

```python
# I return random sequences on the plasmid that are not orivs that way i can have non-oriv in my data for training/testing
negative_records = []
hits = blast[(blast["pident"] >= 95)]
for record in SeqIO.parse("/Users/amamu656/Desktop/GENESIS/blast/plasmids.fasta", "fasta"):
    seq_length = len(record.seq)

    # Find hits for this plasmid
    plasmid_hits = hits[hits["sseqid"] == record.id]

    # Mask oriV regions
    mask = [True] * seq_length
    for _, hit in plasmid_hits.iterrows():
        start = min(hit["sstart"], hit["send"]) - 1
        end = max(hit["sstart"], hit["send"])
        for i in range(start, end):
            mask[i] = False

    # Extract negative regions
    i = 0
    while i < seq_length:
        if mask[i]:
            start = i
            while i < seq_length and mask[i]:
                i += 1
            end = i
            neg_seq = record.seq[start:end]
            if len(neg_seq) >= 20:
                negative_records.append(SeqIO.SeqRecord(neg_seq,
                                                        id=f"{record.id}_neg_{start}_{end}",
                                                        description=""))
        else:
            i += 1

# Save negatives
SeqIO.write(negative_records, "negatives.fasta", "fasta")
```

```python
from Bio import SeqIO
from random import randint
from Bio.SeqRecord import SeqRecord

records = SeqIO.parse("negatives.fasta", "fasta")

fragments = []
frag_counter = 1

for record in records:
    seq_len = len(record.seq)

    if seq_len > 100:  # only fragment if sequence is long enough
        for i in range(10):
            # ensures fragment length is within 100 to 2000 and does not exceed the sequence
            max_frag_len = min(1900, seq_len)
            frag_len = randint(100, max_frag_len)

            start = randint(0, seq_len - frag_len)
            end = start + frag_len

            frag_seq = record.seq[start:end]
            fragment_record = SeqRecord(
                seq=frag_seq,
                id=f"{record.id}_fragment_{frag_counter}",
                name=record.id,
                description=""
            )
            fragments.append(fragment_record)
            frag_counter += 1
    else:
        # keep the whole sequence if it's too short
        fragment_record = SeqRecord(
            seq=record.seq,
            id=record.id,
            name=record.id,
            description=""
        )
        fragments.append(fragment_record)


SeqIO.write(fragments, "fragments.fasta", "fasta")
```

```python
import re
pattern = re.compile(r"^([A-Z]{2}_[A-Z0-9]+\.\d+)")
rows = []
for record in SeqIO.parse("fragments_dedup.fasta", "fasta"):
    match = pattern.match(record.name)
    plasmid_id = match.group(1) if match else None

    rows.append({
        "oriv_sequence": str(record.seq),
        "type_seq": "fragment",
        "plasmid_id": plasmid_id
    })

fragments_df = pd.DataFrame(rows)
```
